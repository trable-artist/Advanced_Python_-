```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
data = {
    'Name': ['Temirlan', 'Adam', 'Arlan', 'Bobr', 'Rob'],
    'Age': [100, 200, 5, 52, 16],
    'Grade': [80, 0, 12, 13, 100]
}
df = pd.DataFrame(data)

plt.plot(df['Name'],df['Grade'],marker='o', color='r')
plt.xlabel('Name')
plt.ylabel('Grade')
plt.title("Competition")
plt.show()

plt.bar(df['Name'], df['Grade'], color='skyblue')        
plt.xlabel('Student Name')                               
plt.ylabel('Score')
plt.title("competition 2")
plt.show()

```
