```python
import pandas as pd
```

```python
data={'Name' : ['Temirlan','Adam','Arlan','Bobr','Rob'],
     'Age' : [100,200,5,52,16],
     'Grade' : [80,0,12,13,100]
     }

df=pd.DataFrame(data)
print(df)

print(f"first 3 rows :\n{df.head(3)}")

print(f"description:\n {df.describe()}")

print(f"avg grade: {df['Grade'].mean()}"

student_age20=df[df['Age']>20]
student_grd80=df[df['Grade']>80]

print(f"studen older 20 :{student_age20}\n")
print(f"studen who got more than 80: {student_grd80}\n")
```
