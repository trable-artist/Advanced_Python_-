```python
import pandas as pd
```

```python
data={'Name' : ['Temirlan','Adam','Arlan','Bobr','Rob'],
     'Age' : [100,200,5,52,16],
     'Grade' : [80,0,12,13,100]
     }

df=pd.DataFrame(data)
print(df)
print("first 3 rows:\n")
print(f"{df.head(3)}")
print("descripition:\n")
print(f"{df.describe()}")
print("avg grade:\n")
print("{df['Grade'].mean()}")

student_age20=df[df['Age']>20]
student_grd80=df[df['Grade']>80]

print("student who older 20\n")
print(f"{student_age20}\n")
print(f"student who got mare than 80:\n")
print(f"{student_grd80}")
```
